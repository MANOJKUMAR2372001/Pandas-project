```python
#custamise data in matplotlib 
import seaborn as sns                                
import matplotlib.pyplot as plt
sns.barplot(x="Vehicle_Make",y="Year",data=df,estimator=sum)
plt.xlabel("Make")
plt.ylabel("Month")
plt.show()
```

```python
#linechart-tranding
df_new=df.groupby("Year")["Price_(USD)"].sum()
plt.figure(figsize=(10,5))
unique_years=df_new.index
df_new.plot(kind="line",marker="o",color="red",linestyle="--")
plt.xticks(unique_years)
plt.show()
```

```python
#pie chart
df_new=df.groupby("Year")["Price_(USD)"].sum()
plt.figure(figsize=(12,6))
df_new.plot(kind='pie',labels=df_new.index,autopct="%1.1f%%")
plt.show()
```

```python
#pie chart
df_new=df.groupby("Vehicle_Make")["Year"].sum()
plt.figure(figsize=(12,6))
df_new.plot(kind='pie',labels=df_new.index,autopct="%1.1f%%")
plt.show()
```
